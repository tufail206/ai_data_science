```python
print('''
 ----------------------------------------------------
                COLLEGE PORTAL
----------------------------------------------------
''')

students = []

while True:  # loop until user chooses to quit
    print('''
press 1 : add student 
press 2 : view students 
press 3 : delete student 
press 4 : quit
''')

    choice = int(input("Enter your choice: "))

    if choice == 1:
        name = input("Enter the name: ")
        fatherName = input("Enter father's name: ")
        age = input("Enter age: ")
        students.append({
            "name": name,
            "fatherName": fatherName,
            "age": age
        })
        print(f'{name} has been added successfully.')

    elif choice == 2:
        if not students:
            print("No students found.")
        for std in students:
            print(f'Name: {std["name"]}, Father: {std["fatherName"]}, Age: {std["age"]}')

    elif choice == 3:
        name_to_delete = input("Enter the name of the student to delete: ")
        students = [std for std in students if std["name"].lower() != name_to_delete.lower()]
        print(f'{name_to_delete} has been deleted if existed.')

    elif choice == 4:
        print("Exiting portal. Goodbye!")
        break  # quit the loop

    else:
        print("Invalid choice. Try again.")
```
