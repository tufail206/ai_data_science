###### exception handling

```python
num1=int(input("enter first number"))
num2=int(input("enter second number"))


print(num1/num2)
```

```python
# exception handle

try:
    num1=int(input("enter first number"))
    num2=int(input("enter second number"))
    print(num1/num2) 
except ZeroDivisionError:
   print("not possible to divide")
```

```python
# if valueError as well
try:
    num1=int(input("enter first number"))
    num2=int(input("enter second number"))
    print(num1/num2) 
except ZeroDivisionError:
   print("not possible to divide by 0")
except ValueError:
    print("invalid value ")
```

```python
# if type error
try:
    num1=int(input("enter first number"))
    num2=input("enter second number")
    print(num1-num2) 
except TypeError:
    print("invalid value data type ")
```

```python
# simple and short form of the error handle

try:
    num1=int(input("enter first number"))
    num2=input("enter second number")
    print(num1-num2) 
except Exception:
    print("error handled at server side")
    
```

```python
# throw the real error message 

# simple and short form of the error handle

try:
    num1=int(input("enter first number"))
    num2=input("enter second number")
    print(num1-num2) 
except Exception as error:
    print(f"error handled at server side : {error}")
```
