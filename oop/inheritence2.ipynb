```python
class Animal():
    def __init__(self, typ, legs, color, sound):
        # instance attributes
        self.typ = typ
        self._legs = legs
        self.color = color
        self.sound = sound
    # instance methods
    def get_legs(self):
        return f"I have {self._legs} legs"
    def set_legs(self,new_legs):
        if new_legs>4:
            return "I cant have more than 4 legs"
        self._legs = new_legs
        return f"Legs updated successfully. New legs are {self._legs}"
    
    def eat(self):
        if self.typ =="carnivore":
            return "Animal is eating meat"
        else:
            return "Animal is eating grass"
    def make_sound(self):
        return self.sound
```

```python
class Lion(Animal):
    def __init__(self,typ,legs,color,sound,character,speed,location):
        #   this super will initialize the parent method and pass the value , the lion initializer combine the parent and his own initializer  
          super().__init__(typ,legs,color,sound)
          self.character=character
          self.speed=speed
          self.location=location
    def hunt(self):
         return f"our {self.color }lion of {self._legs} leg hunts the animal"
```
