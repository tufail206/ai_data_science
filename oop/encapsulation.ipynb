##### Encapsulation

```python
class Animal():
    #to make the attributes protect to access from outside  in encapsulation we use _ before the variable
    def __init__(self,typ,legs,color,sound):
        self.typ=typ
        self._legs=legs
        self.color=color
        self.sound=sound
    def getLeg(self):
        return f"you have {self._legs}"
    def setLeg(self,newLeg):
        if newLeg>4:
            return f"its impossible to have more then 4 legs "
        self._legs=newLeg
        return f"you have updated you legs to  {self._legs}"

```

```python
dog=Animal("kutta",4,"black","wooow wowoff wooof")

dog.getLeg()

dog.setLeg(5)
```
