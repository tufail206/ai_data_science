```python
class Car():
    def __init__(self, make, model, year, engine_cc, color, transmission):
        # Attrbibutes / Characterstics / Properties
        self.make = make
        self.model = model
        self.year = year
        self.engine_cc = engine_cc
        self.color = color
        self.transmission = transmission  
    # Behaviors
    def start(self):
        return f"{self.color} {self.make} {self.model} {self.year} is starting!!!!!!!"
    def fill_tank(self):
        return f"""
                     {self.color}
                     {self.make}
                     {self.model}
                     {self.year}
                    is filling gasoline!"""
```

```python
class ElectricCar(Car):
     pass
```
