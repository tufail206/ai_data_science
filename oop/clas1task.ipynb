```python
class Student():
    def __init__(self,name,age,roll,course,batch):
        self.name=name
        self.roll=roll
        self.age=age
        self.course=course
        self.batch=batch
    def attend_class(self):
        return f"{self.name} is attending {self.course} class"
    def take_exam(self):
        return f"{self.name} is attending {self.course} exam"
    def submitFee(self):
        return f"{self.name} is submitting fee"
    
        
```
