```python
class Files():
    def __init__(self,path):
        self.path=path
    def read_file(self):
        with open(self.path) as file:
            return file.read()
```

```python
file1=Files("data.txt")
```

```python
file1.read_file()
```

```python
class Shop:
    def __init__(self, items: list[dict]):
        self.items = items   

    def displayItem(self):
        for item in self.items:
            print(item["item_name"])

    def userInterface(self):
        print("These are the items available:")
        self.displayItem()

        while True:
            product = input("Enter the item (q to quit): ").lower().strip()

            if product == "q":
                break

            found = False

            for item in self.items:
                if item["item_name"].lower() == product:
                    print("Price:", item["price"])
                    found = True
                    break

            if not found:
                print("No item found")
```

```python
items=[{"item_name":"Soap","price":234}]
```

```python
shop1=Shop(items=items)
```

```python
shop1.displayItem()
```

```python
shop1.userInterface()
```
