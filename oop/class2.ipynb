```python

class Animal():

   def __init__(self,typ,legs,color,sound):
    self.type=typ
    self.legs=legs
    self.color=color
    self.sound=sound
dog=Animal("carnivore",4,"white","Wowowowow")
dog.legs

# //so its easyly can modify like 
dog.legs=6 
# for this we should prevent the data hide the data by encapsulation

# Encapsulation 
# as object can modify the data easyly so we will stop the direct access
```
