```python
class Bank():
    def __init__(self,name,cnic,initial_deposit,account_number):
        self.name=name
        self.cnic=cnic
        # __ is use to make varible private
        self.__balance=initial_deposit
        self.account_number=account_number
    def getBalance(self):
        return f'your balance is {self.__balance}'
    def deposit(self,amount):
        self.__balance+=amount
        return f"""{amount} has been deposit successfully"""
    
```
