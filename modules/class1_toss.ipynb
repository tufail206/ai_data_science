```python
import random

while True:
    toss = input("Coin flip (head/tail) or q to quit: ").lower().strip()

    if toss == "q":
        break

    if toss not in ["head", "tail"]:
        print("Invalid input! Please enter head or tail.")
        continue

    coin = random.choice(["head", "tail"])
    print("Coin result:", coin)

    if toss == coin:
        print("Pakistan have won the toss ðŸ‡µðŸ‡°")
    else:
        print("Pakistan lost the toss")
```
