What is the os module in Python?

The os module allows Python programs to interact with the operating system.

In simple words:

It helps Python talk to your computer’s OS (Windows, Linux, macOS).

Why do we use the os module?

We use os to:

Work with files and folders

Get system information

Run OS-level commands

Handle paths and directories

Access environment variables

Common uses of the os module
1. Importing the module
import os

2. Get current working directory
print(os.getcwd())

3. List files in a directory
print(os.listdir())

4. Create and remove folders
os.mkdir("test_folder")
os.rmdir("test_folder")

5. Check if a file or folder exists
os.path.exists("file.txt")

6. Join paths safely (very important!)
path = os.path.join("folder", "file.txt")

7. Run system commands
os.system("echo Hello")

Is os an Operating System?

❌ No, os is not an operating system
✅ It is a Python module that interacts with the operating system
