module : 
A module is simply a file that contains Python code — like variables, functions, or classes — that you can reuse in other programs.

Why do we use modules?

Modules help you write better, cleaner, and reusable code.

Main reasons:

Code reusability
Write code once, use it many times.

Better organization
Instead of one huge file, you split code into logical parts.

Easier maintenance
Bugs are easier to fix when code is separated.

Avoid name conflicts
Variables and functions stay neatly inside their modules.

Access built-in features
Python comes with tons of ready-made modules.

```python
import math
```

```python
math.sin(4)
```

```python
math.log(5)
```
