modes in python 
'r':read
'w':write
'a':append
'w+':write and read
'r+':read and write
'rb'"read binary
'wb':write binary

```python
# to append also the text
with open("file.txt","a") as f:
    f.write("  high demand")
```

```python
# to append in next line we can use \n
with open("file.txt","a") as f:
    f.write("\ngood for Ai")
```
