```python
import random
```

```python
all_account=[]
def open_account(name:str,cnic:str,initial_deposit:int):
    if len(cnic)>=1:
         all_account.append({'name':name,'cnic':cnic,'balance':initial_deposit,'pin':random.randint(1001,9999),'account_number':random.randint(10111,90111)})
         return f"{name} your account is created successfully"
    else:
         return "invalid cnic"
    
```

```python
def deposite(account_number:int,amount:int):
    for account in all_account:
        if account['account_number']==account_number:
            account['balance']+=amount
            return f"{amount} deposit successfully.Your account has {account['balance']}"
        else:
            return f"invalid account number"
```

```python
def withdraw(account_number:int,pin:int,amount:int):
    for account in all_account:
        if account['account_number']==account_number:
            if account['pin']==pin:
                if account['balance']>amount:
                    account['balance']-=amount
                    return f"{amount} withdraw successfully "
                else:
                      return f"{amount} invalid successfully "
            else:
                return f"invalid pin"
        else:
             return f"no account found"
    
         
```
