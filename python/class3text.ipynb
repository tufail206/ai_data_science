```python
# list of dictionary
Shop=[
    {'product_name':"Soap",'price':340,'quantity':50 },
    {'product_name':"Surf",'price':240,'quantity':40 },
    {'product_name':"Oil",'price':230,'quantity':20},     {'product_name':"Socks",'price':90,'quantity':45},     {'product_name':"Shampoo",'price':195,'quantity':50}, {'product_name':"Gee",'price':690,'quantity':34},
    {'product_name':"Egg",'price':380,'quantity':54},
    {'product_name':"Bread",'price':230,'quantity':42}
]


```

```python
cart=[]
while True:
     purchase={}
     items=input("what you want to buy")
     if items=="q":
          print("thank you for stay with us")
          break
     for prod in Shop:
          if prod['product_name']==items:
               quantity=int(input("how much quantity"))
               if prod["quantity"]>=quantity:
                    amount=quantity*prod["price"]
                    purchase["amount"]=amount
                    purchase["item"]=items
                    purchase["quantity"]=quantity

                    cart.append(purchase)
                    prod["quantity"]-=quantity
                    break
               else:
                    print(f"{quantity} is not available")

     else:
          print(f"{items} not found ")
```
