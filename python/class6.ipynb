```python
# random.randint(1001,9999) to create new random number 
all_account=[]
import random
```

```python
def open_account(cnic,title,initial_deposite):
       acount={'cnic'              :cnic,
               'title'             :title,
               'balance'           :initial_deposite,
               'account_number'    :random.randint(1001,9999),
               'pin'               :random.randint(1001,9999)
               }
       all_account.append(acount)
       return  (
    f"Your account has been created successfully\n"
    f"Your account number is {acount['account_number']}\n"
    f"Your pin number is {acount['pin']}"
)
```

```python
open_account("123","adnan",6000)
```

```python
def check_balance(account,pin):
    for acc in all_account:
        if  acc['account_number']==account:
            if acc['pin']==pin:
                 return acc['balance']
            else:
                 print("invalid pin")
                 break # because dont want to loop on other account 
    else:  # i append else with for loop because if acount not found every time print invalid account so that's why 
            print("invalid account number")

```

```python
check_balance(7346,1726)
```

```python
def withdraw_balance(account_number,pin,amount):
    for acc in all_account:
        if acc['account_number']==account_number:
            if acc['pin']==pin:
             if acc['balance']>=amount:
                  acc['balance']-=amount
                  return f"{amount} is withdraw .Your current balance is {acc['balance']}"
           
             else:
                 return "invalid amount" 
            
    else:
        return "invalid account number"

```

```python
withdraw_balance(2793,4418,100)
```

```python
def deposit(account,amt):
    for acc in all_account:
        if acc["account_number"]==account:
            acc["balance"]+=amt
            return f"Rs:{amt} is deposited successfully.Your current balance is {acc["balance"]}"
    else:
        return "invalid acount number"
```
