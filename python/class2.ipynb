#product dictionary 

```python
product={'soap':120,'shampo':230,'surf':300,'tooth_paste':120,'tooth_burs':400}
```

```python
print('All in one shop')
print(product.keys())
```

```python
orders=[]
for i in range(5):
    order=input('what kind of product do you have to byt').lower().strip()
    if order in product.keys():
        productPrice=product[order]
        orders.append(productPrice)
    else:
        print("product not found")
total=sum(orders)
print(total)
```
