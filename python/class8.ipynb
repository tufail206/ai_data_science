```python
import time 
async def myFun():
      print("myFun start")
      time.sleep(2)  #after two second
      print("myFun end")
startTime=time.time()
await myFun()
endTime=time.time()
print(endTime-startTime)
```

```python
# decorator in python enhances the function

def khanDecorator(func):
    def wrapper(a):
        return func(a) + " khan"
    return wrapper

@khanDecorator
def nameReturn(name):
    return name

print(nameReturn("adnan"))
```
