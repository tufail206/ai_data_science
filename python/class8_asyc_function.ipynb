##### async def function :
         async def  func():    it will run on the background and will run after completion
         


```python
import time 
def myFun():
      print("myFun start")
      print("hello")
      print("hello")
      print("hello")
      print("hello")
      print("hello")
      print("hello")
      print("hello")
      print("myFun end")
startTime=time.time()
myFun()
endTime=time.time()
print(endTime-startTime)
```
