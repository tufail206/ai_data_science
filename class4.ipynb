##### How to handle more argument in a function 

```python
# parameters :  default parameter 
def saySalam(name="zain"):
    print(f"salam {name}")
saySalam("adnan")
```

```python
def FullName(fName,lName,mName=""):
    print(f"hello {fName} {mName} {lName} how are your ?")
FullName("tufail","zain","abbas")
```

###### arguments


```python
# positional arguments
# sequence wise 

def FullName(fName,lName,mName=""):
    print(f"hello {fName} {mName} {lName} how are your ?")
FullName("tufail","zain","abbas")
```

```python
# keyword argument

def FullName(fName,lName,mName=""):
    print(f"hello {fName} {mName} {lName} how are your ?")
FullName(fName="tufail",lName="zain",mName="abbas")

# in keyword argument there might be both keyword and positonal
fN=FullName("tufail" ,lName="zain",mName="abbas")  
print(fN)
```

```python
def add(*nums):
    return sum(nums)
x=add(23,4,5)
print(x)
```
