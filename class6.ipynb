```python
# random.randint(1001,9999) to create new random number 
all_account=[]
import random
```

```python
def open_account(cnic,title,initial_deposite):
       acount={'cnic'              :cnic,
               'title'             :title,
               'balance'           :initial_deposite,
               'account_number'    :random.randint(1001,9999),
               'pin'               :random.randint(1001,9999)
               }
       all_account.append(acount)
       return f"""your account has been created successfully
                  your account number is {acount['account_number']}
                  your pin number is {acount['pin']}"""
```

```python
open_account("999","adnan",6000)
```

```python
def check_balance(account,pin):
    for acc in all_account:
        if  acc['account_number']==account:
            if acc['pin']==pin:
                 return acc['balance']
            else:
                 print("invalid pin")
                 break # because dont want to loop on other account 
    else:  # i append else with for loop because if acount not found every time print invalid account so that's way 
            print("invalid account number")

```

```python
check_balance(7346,1726)
```
